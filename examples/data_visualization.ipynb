This is an example of the type of data visualization that will allow us to visualize the relationships between multiple variables.

```python
import pandas as pd
import seaborn as sns
```

Seaborn is just a nicer plotting library, since matplotlib and the traditional MATLAB plots are quite obtuse and unappealing (IMO).

```python
# load the data
game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
games
```

```python
game_data
```

For this example, I will take a look at the relationship between the number of turnovers and point differential.

First, notice that the score data and the game statistics are in different datasets. We need to merge the two tables together to access both at the same time. You can also save this combined dataset to file so we don't have to merge every time.

We will accomplish this using the `merge` function from Pandas. This merges two dataframes together by linking on the columns specified. Here, we use 'id' to merge the two games together, since the game ID forms the relationship between the two tables. The tables are joined and each entry in the second table are moved to their respective location in the first table. Notice too how the 2021 games that are yet to be played in the `games.csv` dataset are not joined into the combined data table since there is no game statistic data on games that haven't yet been played. The `merge` function auto drops these entries.

Further, we validate that there is 1-1 mapping between both tables so that we aren't duplicating entries, using the parameter `validate` and setting it to `"one_to_one"`.

```python
combined_data = games.merge(game_data, left_on="id", right_on="id", validate="one_to_one")
combined_data
```

Now we will generate the Point differential. Here, we subtract the scores scored by the away team from the home team.

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]


combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
combined_data["point_differential"]
```

Here we use more complex features of both Python and Pandas:
- The `apply` function takes a Python function as an input, and executes that function for elements in the DataFrame, and returns the modified DataFrame. Here give `apply` the function `point_diff` as an input
- We tell `apply` to act on `axis=1`, which means to apply the function `point_diff` to all rows. If we were instead to specify `axis=0`, `point_diff` would instead be applied to all column

Now we have point differential, we will plot it against turnover rate for both home and away teams. As turnovers occur in discrete counts, it makes sense here to plot it as a bar chart. If you are working with continuous data, then a scatter plot makes more sense

```python
sns.barplot(
    data=combined_data, x="home.turnovers", y="point_differential"
).set_title("Home Team Turnovers vs Point Differential")
```

From this data, we can see that the more turnovers a team has, the lower the average point differential that game has. This makes sense, as if the home team makes many turnovers, they lose more opportunities to score points, and so they hence are more likely to lose (and therefore the point differential for this game is lower). We should expect the inverse relationship for the turnovers the away team makes:

```python
sns.barplot(
    data=combined_data, x="away.turnovers", y="point_differential"
).set_title("Away Team Turnovers vs Point Differential")
```

As expected, the more turnovers the away team has, the greater is the average point differential, meaning that the away team has much fewer points than the home team, indicating the home team wins more often.

We can do one more analysis: look at how the difference in the number of turnovers impacts the point differential:

```python
def turnover_diff(row):
    return row["home.turnovers"] - row["away.turnovers"]

combined_data["turnover_differential"] = combined_data.apply(turnover_diff, axis=1)
sns.barplot(
    data=combined_data, x="turnover_differential", y="point_differential"
).set_title("Turnover Differential vs Point Differential")
```

This shows that when the away team makes many more turnovers than the home team, the home team will gain many more points than the away team, and conversely, when the home team makes many more turnovers than the away team, the away team will gain many more points than the home team. 

This relationship, while not fully statistically justified, lets us know informally that historical turnover differential might be good indicate of which team is more likely to win.
